```python
df['active_cases'] = df['Confirmed'] - df['Recovered'] - df['Deaths']
df.tail()
```

```python
df.head(50)
```

```python
df = df.drop(columns = ['State'])
```

```python
df = pd.pivot_table(df , values = ['Confirmed','Deaths', 'Recovered','active_cases'],index = 'Region', aggfunc = max)
```

```python
df['Recovery_Rate'] = df['Recovered'] * 100 / df['Confirmed']
```

```python
df['Mortality_Rate'] = df['Deaths'] * 100 / df['Confirmed'] 
```

```python
df.style.background_gradient(cmap = 'binary')
```

```python
df['zscore'] = (df.Confirmed - df.Confirmed.mean())/df.Confirmed.std()
df.head(5)
```
